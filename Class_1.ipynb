f- string formatting types-

variable {x}
expression {x+y}
precision {x:.2f}
width {x:8}
alignment {x:^8}
padding {x:*^8}
zero padding {x:05}
percentage {X:2%}
commas {x:,}


```python
print(f"the sum of the numbers is {a+b}")
price=3600.78965
print(f"{price:.2f}") # precision

x=42
print(f"{x:7}") #width

p="python"
print(f"|{p:<15}|") # left alignment 
print(f"|{p:>15}|") # right alignment
print(f"|{p:^15}|") # center alignment 

print(f"{42:*<6}") #padding
print(f"{42:*>6}")
print(f"{42:*^6}")

print(f"{7:03}") # zero padding
print(f"{123:06}")

value=3.5
print(f"{value:07.2f}") # zero padding precision

ratio=0.7865
print(f"{ratio: .2%}") # percentage formatting

salary=12500000
print(f"{salary:,}")
print(f"{salary:,.2f}") # cpmmas(thousands separator)
```
